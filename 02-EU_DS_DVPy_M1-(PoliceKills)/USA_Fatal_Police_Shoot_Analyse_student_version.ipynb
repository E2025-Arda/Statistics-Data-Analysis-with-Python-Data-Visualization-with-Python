## 2. Most common 15 Name or Surname of killed people 

```python
kill.name.value_counts().head(15)
```

```python
kill.shape
```

```python
kill.head()
```

```python
# Problem: TK TK is most probably not a name or surname. We will clean them.
pairs = kill.name[kill.name != 'TK TK'].str.split()
pairs.head()
```

```python
type(pairs)
```

```python
pairs = kill.name[kill.name != 'TK TK'].str.split()
langlist =[]
for i in pairs:
    langlist.extend(i)
pl=pd.Series(langlist).value_counts().head(15)
pl
```

**INSTRUCTION : Plot Most common 15 Name or Surname of killed people** 

```python
pl.index
```

```python
fig, ax = plt.subplots(figsize=(12,9))

g=ax=sns.barplot(x=pl.index, y= pl.values, ax=ax)

ax.set_ylabel('Frequency')
ax.set_xlabel('names or surname of killed people')
ax.set_title('most common 15 name or surname of killed people') 
ax.set_xticklabels(pl.index, rotation=45)
for p in g.patches:
    g.annotate((p.get_height()), (p.get_x()+0.2, p.get_height()+0.5))

plt.show()
```
